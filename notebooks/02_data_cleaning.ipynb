```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/bluejays_batting_2024_raw.csv')
```

```python
df[df['Rk'].isna()][['Player','Rk','Age','Pos']]
```

```python
df_clean = df[df['Rk'].notna()].copy()
```

```python
df_clean['Player'] = (
    df_clean['Player']
    .str.replace(r'[\*\#]', '', regex=True)
    .str.strip()
)
```

```python
df_clean = df_clean.drop(columns=['Pos.1'])
```

```python
df_clean['Rk'] = df_clean['Rk'].astype(int)
df_clean['Age'] = df_clean['Age'].astype(int)
```

```python
df_clean.to_csv('tor_hitters_2025_clean.csv', index=False)
```
