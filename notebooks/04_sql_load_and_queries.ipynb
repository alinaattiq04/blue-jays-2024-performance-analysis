```python
import pandas as pd
import sqlite3

df = pd.read_csv('/content/bluejays_batting_2024_clean.csv')

conn = sqlite3.connect('bluejays.db')

df.to_sql('hitters_2025', conn, if_exists='replace', index=False)

df.head()
```

```python
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
```

```python
query = """
SELECT Player, WAR, PA, [OPS+]
FROM hitters_2025
ORDER BY WAR DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```
